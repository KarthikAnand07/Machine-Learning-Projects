Annual Rainfall Trends

```python
# Plotting the annual rainfall trends
plt.figure(figsize=(14, 8))
for subdivision in df['SUBDIVISION'].unique():
    subset = df[df['SUBDIVISION'] == subdivision]
    plt.plot(subset['YEAR'], subset['ANNUAL'], label=subdivision)
plt.title('Annual Rainfall Trends by Subdivision')
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Seasonal Rainfall Analysis

```python
# Calculating average seasonal rainfall for each subdivision
seasonal_avg = df.groupby('SUBDIVISION')[['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']].mean().reset_index()

# Plotting seasonal rainfall
seasonal_avg_melted = pd.melt(seasonal_avg, id_vars=['SUBDIVISION'], value_vars=['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'])
plt.figure(figsize=(14, 8))
sns.barplot(x='SUBDIVISION', y='value', hue='variable', data=seasonal_avg_melted)
plt.title('Average Seasonal Rainfall by Subdivision')
plt.xlabel('Subdivision')
plt.ylabel('Average Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()
```

Yearly Rainfall Anomalies

```python
# Calculating yearly anomalies
df['ANOMALY'] = df.groupby('SUBDIVISION')['ANNUAL'].transform(lambda x: (x - x.mean()) / x.std())

# Plotting anomalies
plt.figure(figsize=(14, 8))
for subdivision in df['SUBDIVISION'].unique():
    subset = df[df['SUBDIVISION'] == subdivision]
    plt.plot(subset['YEAR'], subset['ANOMALY'], label=subdivision)
plt.axhline(0, color='gray', linestyle='--')
plt.title('Yearly Rainfall Anomalies by Subdivision')
plt.xlabel('Year')
plt.ylabel('Anomaly (Standard Deviations)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Comparison of Rainfall Across Subdivisions

```python
# Plotting comparison of annual rainfall
plt.figure(figsize=(14, 8))
sns.boxplot(x='SUBDIVISION', y='ANNUAL', data=df)
plt.title('Comparison of Annual Rainfall Across Subdivisions')
plt.xlabel('Subdivision')
plt.ylabel('Annual Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()
```

Correlation Between Monthly Rainfalls

```python
# Calculating correlation matrix for monthly rainfall
monthly_columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
correlation_matrix = df[monthly_columns].corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Monthly Rainfall')
plt.show()
```

Rainfall Extremes

```python
# Calculating the range of rainfall for each subdivision
df['RANGE'] = df.groupby('SUBDIVISION')['ANNUAL'].transform(lambda x: x.max() - x.min())

# Plotting extreme rainfall variations
extreme_rainfall = df[['SUBDIVISION', 'RANGE']].drop_duplicates().sort_values(by='RANGE', ascending=False)
plt.figure(figsize=(14, 8))
sns.barplot(x='SUBDIVISION', y='RANGE', data=extreme_rainfall)
plt.title('Extreme Rainfall Variations by Subdivision')
plt.xlabel('Subdivision')
plt.ylabel('Range of Annual Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()
```

Impact of Monsoon

```python
# Calculating the percentage of rainfall during the monsoon season
df['Monsoon_Percentage'] = (df['Jun-Sep'] / df['ANNUAL']) * 100

# Plotting the monsoon contribution
plt.figure(figsize=(14, 8))
sns.boxplot(x='SUBDIVISION', y='Monsoon_Percentage', data=df)
plt.title('Monsoon Contribution to Annual Rainfall by Subdivision')
plt.xlabel('Subdivision')
plt.ylabel('Monsoon Rainfall as Percentage of Annual Rainfall')
plt.xticks(rotation=90)
plt.show()
```

Rainfall during Non-Monsoon Seasons

```python
# Plotting non-monsoon seasonal rainfall
non_monsoon_seasonal_avg = df.groupby('SUBDIVISION')[['Jan-Feb', 'Mar-May', 'Oct-Dec']].mean().reset_index()

# Plotting non-monsoon seasonal rainfall
non_monsoon_avg_melted = pd.melt(non_monsoon_seasonal_avg, id_vars=['SUBDIVISION'], value_vars=['Jan-Feb', 'Mar-May', 'Oct-Dec'])
plt.figure(figsize=(14, 8))
sns.barplot(x='SUBDIVISION', y='value', hue='variable', data=non_monsoon_avg_melted)
plt.title('Average Non-Monsoon Seasonal Rainfall by Subdivision')
plt.xlabel('Subdivision')
plt.ylabel('Average Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()
```

Rainfall Variability Analysis

```python
# Calculating the coefficient of variation for each subdivision
df['Coefficient_of_Variation'] = df.groupby('SUBDIVISION')['ANNUAL'].transform(lambda x: x.std() / x.mean() * 100)

# Plotting the variability in annual rainfall
rainfall_variability = df[['SUBDIVISION', 'Coefficient_of_Variation']].drop_duplicates().sort_values(by='Coefficient_of_Variation', ascending=False)
plt.figure(figsize=(14, 8))
sns.barplot(x='SUBDIVISION', y='Coefficient_of_Variation', data=rainfall_variability)
plt.title('Rainfall Variability by Subdivision')
plt.xlabel('Subdivision')
plt.ylabel('Coefficient of Variation (%)')
plt.xticks(rotation=90)
plt.show()
```

Top Rainfall Months

```python
# Calculating average monthly rainfall across all subdivisions
monthly_avg_rainfall = df[monthly_columns].mean().sort_values(ascending=False)

# Plotting the average monthly rainfall
plt.figure(figsize=(14, 8))
sns.barplot(x=monthly_avg_rainfall.index, y=monthly_avg_rainfall.values)
plt.title('Average Monthly Rainfall Across All Subdivisions')
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm)')
plt.show()
```

Changes in Monsoon Rainfall Over Time

```python
# Plotting the trend of monsoon rainfall over time
plt.figure(figsize=(14, 8))
for subdivision in df['SUBDIVISION'].unique():
    subset = df[df['SUBDIVISION'] == subdivision]
    plt.plot(subset['YEAR'], subset['Jun-Sep'], label=subdivision)
plt.title('Trends in Monsoon Rainfall Over Time by Subdivision')
plt.xlabel('Year')
plt.ylabel('Monsoon Rainfall (mm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Rainfall Correlation with Time

```python
# Calculating correlation between annual rainfall and time
correlations = df.groupby('SUBDIVISION').apply(lambda x: x['YEAR'].corr(x['ANNUAL'])).reset_index()
correlations.columns = ['SUBDIVISION', 'Correlation_with_Time']

# Plotting the correlation
plt.figure(figsize=(14, 8))
sns.barplot(x='SUBDIVISION', y='Correlation_with_Time', data=correlations)
plt.title('Correlation Between Annual Rainfall and Time by Subdivision')
plt.xlabel('Subdivision')
plt.ylabel('Correlation with Time')
plt.xticks(rotation=90)
plt.show()
```

Rainfall Contribution by Season

```python
# Calculating the seasonal contributions to annual rainfall
df['Pre-monsoon_Contribution'] = (df['Mar-May'] / df['ANNUAL']) * 100
df['Monsoon_Contribution'] = (df['Jun-Sep'] / df['ANNUAL']) * 100
df['Post-monsoon_Contribution'] = (df['Oct-Dec'] / df['ANNUAL']) * 100
df['Winter_Contribution'] = (df['Jan-Feb'] / df['ANNUAL']) * 100

# Averaging contributions for all subdivisions
seasonal_contributions = df[['Pre-monsoon_Contribution', 'Monsoon_Contribution', 'Post-monsoon_Contribution', 'Winter_Contribution']].mean()

# Plotting the seasonal contributions
plt.figure(figsize=(14, 8))
sns.barplot(x=seasonal_contributions.index, y=seasonal_contributions.values)
plt.title('Seasonal Contributions to Annual Rainfall Across All Subdivisions')
plt.xlabel('Season')
plt.ylabel('Contribution to Annual Rainfall (%)')
plt.show()
```
