### Distribution of User Ratings

```python
plt.figure(figsize=(10, 6))
sns.histplot(amazon['User Rating'], bins=30, kde=True)
plt.title('Distribution of User Ratings')
plt.xlabel('User Rating')
plt.ylabel('Frequency')
plt.show()
```

### Countplot of genres

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Genre', data=amazon, palette='viridis')
plt.title('Count of Books in Each Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

### Scatter plot of User Rating vs. Reviews

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='User Rating', y='Reviews', data=amazon, hue='Genre', palette='Set1')
plt.title('User Rating vs. Reviews')
plt.xlabel('User Rating')
plt.ylabel('Reviews')
plt.show()
```

### Boxplot of Price by Genre

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Genre', y='Price', data=amazon, palette='pastel')
plt.title('Price Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Price')
plt.show()
```

### Yearly trend of bestselling books

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', data=amazon, palette='viridis')
plt.title('Number of Bestselling Books Each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

### Top authors with the most books

```python
top_authors = amazon['Author'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_authors.index, y=top_authors.values, palette='muted')
plt.title('Top 10 Authors with Most Bestselling Books')
plt.xlabel('Author')
plt.ylabel('Number of Books')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Distribution of book prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(amazon['Price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Book Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### Scatter plot of User Rating vs. Price

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='User Rating', y='Price', data=amazon, hue='Genre', palette='Set2')
plt.title('User Rating vs. Price')
plt.xlabel('User Rating')
plt.ylabel('Price')
plt.show()
```

### Correlation matrix to identify relationships between numerical variables

```python
correlation_matrix = amazon[['User Rating', 'Reviews', 'Price']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
